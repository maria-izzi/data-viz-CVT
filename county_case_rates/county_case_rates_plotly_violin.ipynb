```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
```

```python
df= pd.read_csv("county_case_rates.csv")
# fig=px.line(df,y=["March","April","May","June"],hover_data=["County Name"])
fig = px.violin(df, y=["March","April","May","June"],hover_data=["County Name"],labels={'variable':'Month','value':'Case Rates'},title="Change in Case Rates over Time")
fig.show()
# f = go.FigureWidget(fig)
```

```python
fig.write_html("county_case_rates.html")
```
